```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris
```

```python
import seaborn as sns
```

```python
irish=pd.DataFrame(iris.data)
```

```python
irish.columns=iris.feature_names
```

```python
irish['Species']=iris.target
```

```python
irish
```

```python
X=irish.iloc[:,:-1]
Y=irish.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Prepruning
parameter={
    'criterion':['gini','entropy','log_loss'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
Classifier=DecisionTreeClassifier()
cv=GridSearchCV(Classifier,param_grid=parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,Y_train)
```

```python
cv.best_params_
```

```python
Y_predict=cv.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(Y_predict,Y_test)
score
```

```python
classification_report(Y_predict,Y_test)
```

```python
print(classification_report(Y_predict,Y_test))
```
